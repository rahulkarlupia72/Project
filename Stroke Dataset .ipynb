```python
data_copy=my_data.copy()
data_copy.drop(3116,axis=0,inplace=True)
data_copy.reset_index(drop=True,inplace=True)
```

```python
data_copy['gender'].value_counts()
```

```python
data_copy.head(5)
```

```python
features=data_copy.iloc[:,1:-1]
features
```

```python
predictions=data_copy.iloc[:,-1]
predictions
```

```python
data_copy['Residence_type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
features['gender']=lc.fit_transform(features['gender'])

lc1=LabelEncoder()
features['ever_married']=lc1.fit_transform(features['ever_married'])

lc2=LabelEncoder()
features['Residence_type']=lc2.fit_transform(features['Residence_type'])

```

```python
features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[5,9])],remainder='passthrough')
features=ct.fit_transform(features)
```

```python
features
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
features=sc.fit_transform(features)
```

```python
features
```

```python
features.shape
```

```python
predictions=predictions.values
predictions
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,predictions,test_size=0.2,random_state=45,shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(criterion='entropy',n_estimators=100,class_weight='balanced',max_depth=9,random_state=45,ccp_alpha=0.001)
rc.fit(X_train,y_train)
y_pred_train=rc.predict(X_train)
y_pred=rc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Data")
print('Accuracy of Training data',accuracy_score(y_train,y_pred_train))
print("Testing Data")
print('Accuracy of Testing Data ',accuracy_score(y_test,y_pred))
```
